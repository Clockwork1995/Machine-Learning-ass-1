### 5.2
##### Using R (with no use of special libraries), implement analgorithm that you derived in Step I. The implementation is straightforward as you are allowed to use the code examples provided.

```r
# Make prediction given inputs and weights
predict = function(Phi, w){
    return(Phi%*%w)
}

# Error function to calculate errors (Implements L2-Regularizer)
error = function (Phi, w, t, lambda){
    return(.5 * (mean((predict(Phi, w) - t)^2)) + .5 * lambda * w %*% w)
}

sgd_update_weight = function(W_prev, x, y_true, y_pred, lambda, eta) {
   grad = (lambda * W_prev) - ((y_true-y_pred) * x)
   return(W_prev - eta * grad)
}
```

```r
sgd = function(train_data, train_label, lambda, eta, epsilon, max_epoch) {
    train_length = nrow(train_data)
    
    tau_max = max_epoch * train_length
    
    # Initialize train data with additional column for bias 
    Phi = as.matrix(cbind('x0'=1, train_data))
    
    # Intialize weights
    W = matrix( ,nrow=tau_max, ncol=ncol(Phi)) 
    W[1,] = runif(ncol(Phi))
    
    # Intialize tau -  Iteration counter
    tau = 1 # counter
    # Track errors
    errors = data.frame('tau'=1:tau_max)
    errors[tau,] = error(Phi, W[tau,], train_label[,1], lambda)
    while (tau <= tau_max) {
        # Breaking condition for training
        if(errors[tau,] <= epsilon) {break}
        train_sequence = sample(1:train_length, train_length, replace = FALSE)
    
        # loop over each datapoint
        for (i in train_sequence) {
            # increment the counter
            tau = tau + 1
            if (tau > tau_max) {break}
            # make the weight update
            prediction = predict(Phi[i,], W[tau-1,])
            W[tau,] = sgd_update_weight(W[tau-1,], Phi[i,], train_label[i,], prediction[1,], lambda, eta)
            
            # keep track of the objective funtion
            errors[tau,1] = error(Phi, W[tau,],train_label[,1], lambda)
       }
    }
    return (list("error"=errors, "weights"=W))
}
```

```r
calculate_errors = function(train_data, train_label, test_data, test_label, W) {
    train_data = as.matrix(cbind('x0'=1, train_data))
    test_data = as.matrix(cbind('x0'=1, test_data))
    
    n_weights = nrow(W)
    errors = data.frame("train"=rep(0, n_weights), "test"=rep(0, n_weights))
    for (tau in 1:n_weights) {
        errors[tau,"train"] = error(train_data, W[tau,],train_label[,1], 0)
        errors[tau,"test"] = error(test_data, W[tau,],test_label[,1], 0)
  }
  return (errors)
}
```

```r
# Hyper-Parameters
eta = 0.01 # Learning rate
epsilon = 0.001 # Terminating condition
max_epoch = 20 # Number of iterations (Epochs)
```
